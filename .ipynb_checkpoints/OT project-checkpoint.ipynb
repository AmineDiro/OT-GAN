```python
# Parametres

batch_size=64 # selon les auteurs mais on va pas utiliser ça
ginput_dim = 100

#val_batch_size=128
n_epochs=20
lr=3e-4
beta1= 0.5
beta2=0.999
epsilon=1
n_gen=3 #nupmber of generator updates / discriminator update

output_path_imgs ="./generated_image"
save_epoch = 2
sample_interval = 2

##Saved model path

#save_path = 'model/model_mnist.pth'
#model_path = max(save_path, key=os.path.getctime)
#state_dict = torch.load(model_path,map_location='cpu')
#gmodel.load_state_dict(state_dict['generator'])
```

```python
use_cuda =True
# Init device 
print("CUDA Available: ",torch.cuda.is_available())
device = torch.device("cuda" if (use_cuda and torch.cuda.is_available()) else "cpu")
```

```python
#Create Generator
gmodel = Generator()
# Create the discriminator.
dmodel = Discriminator()
```

```python
dataloader = load_CIFAR10()
```

```python
from matplotlib.pyplot import figure

figure(figsize=(10, 10), dpi=80)

plt.imshow(np.transpose(vutils.make_grid(
    d1[ : 64], padding=2, normalize=True).cpu(), (1, 2, 0)))
```

```python
optimizer_D = optim.Adam(dmodel.parameters(), lr=lr, betas=[beta1,beta2])
optimizer_G = optim.Adam(gmodel.parameters(), lr=lr,betas=[beta1,beta2])

mone = -torch.tensor(1, dtype=torch.float).to(device)

for epoch in range(n_epochs):
    desc = "[Epoch {}]".format(epoch)
    torch.cuda.empty_cache()
    tepoch = tqdm(enumerate(dataloader), desc=desc)
    for i, ((im1,_), (im2,_)) in tepoch:
        optimizer_D.zero_grad()
        optimizer_G.zero_grad()
        #Real images batch to GPU
        im1 , im2 = im1.to(device) , im2.to(device)
        #Generate batch noise
        noise = ((torch.rand([batch_size, ginput_dim])*-2)+1).to(device)
        prime_noise = ((torch.rand([batch_size, ginput_dim])*-2)+1).to(device)
        #Generated images :
        gen_img = Variable(gmodel(noise),requires_grad=False)
        gen_img_prime = Variable(gmodel(prime_noise),requires_grad=False)
        
        # Encoded Generated images
        Y , Yprime = dmodel(gen_img), dmodel(gen_img_prime)
        
        
        # Encoded Real images
        X = dmodel(im1)
        Xprime = dmodel(im2)

        loss, loss_generator, loss_critic = loss_func(X, Xprime, Y, Yprime)
         
            
        #loss.backward()
        #optimizer_G.step()

        if i+1 % n_gen == 0:
            # Update critic once every 3 generator updates
            loss.backward(mone)
            optimizer_D.step()
        else:
            loss.backward()
            optimizer_G.step()
        #descriptio tqdm
        
        tepoch.set_postfix(
                loss_generator=loss_generator.item(), loss_critic=loss_critic.item()
            )
        #print("[Epoch {}/{}] [G loss: {}] [D loss: {}]".format(epoch, 
        #                                                        n_epochs,                         
        #                                                        loss_generator.item(),
        #                                                        loss_critic.item()))
    #if epoch%sample_interval==0:
        save_image(gen_img.data[:64], os.path.join(output_path_imgs, "epoch_{}.png".format(epoch)), 
                       nrow=8, normalize=True)
    
    if epoch % save_epoch == 0:
        torch.save({
            'generator' : gmodel.state_dict(),
            'critic' : dmodel.state_dict(),
            'optimizer_G' : optimizer_G.state_dict(),
            'optimizerD' : optimizer_D.state_dict(),
            }, 'model/model_epoch_{}.pth'.format(epoch))
    
```

```python
import os
import glob
import matplotlib.pyplot as plt
import matplotlib.animation as animation

import numpy as np
import torch
import torchvision
from torchvision import datasets, transforms
from torch.utils.data import SubsetRandomSampler, Subset
from torchvision.utils import save_image
import torch.nn as nn
from torch import Tensor
import torch.nn.functional as F
import torch.optim as optim

import torchvision.utils as vutils
from torch.utils.data import DataLoader
from torch.autograd import Variable
from tqdm import tqdm

from SinkhornDistance import loss_func
from dataset import load_MNIST, load_CIFAR10
from Generator import Generator
from Discriminator import Discriminator


from utils import *
#import ot
%load_ext autoreload
%autoreload 2
%reload_ext autoreload
```

```python
from InceptionScore import inception_score_training
from torchvision.models.inception import inception_v3
```

```python
#Create Generator
gmodel = Generator()

gmodel = gmodel.to(device)
```

```python
inception_model = inception_v3(pretrained=True, transform_input=False).to(device)
inception_model.eval()
```

```python
N=100
up = nn.Upsample(size=(299, 299),  align_corners=False ).to(device)
batch_size=32
preds = np.zeros((batch_size*N, 1000))
```

```python
with torch.no_grad():
    for i  in range(N):
        gmodel.eval()
        z_random = (-2 * torch.rand((batch_size, 100)) + 1).to(device)
        samples = gmodel(z_random)
        pred = F.softmax(inception_model(up(samples)),dim=1).data.cpu().numpy()
        preds[i*batch_size:i*batch_size + batch_size] = pred
        del z_random
        torch.cuda.empty_cache()
```

```python
from scipy.stats import entropy

splits=10
# Now compute the mean kl-div
split_scores = []

for k in range(splits):
    part = preds[k * (N // splits): (k+1) * (N // splits), :]
    py = np.mean(part, axis=0)
    scores = []
    for i in range(part.shape[0]):
        pyx = part[i, :]
        scores.append(entropy(pyx, py))
    split_scores.append(np.exp(np.mean(scores)))
```
